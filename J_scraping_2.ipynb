```python
import pandas as pd 
from bs4 import BeautifulSoup as bs
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import nbconvert
import requests
import time
from splinter import Browser
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
rt_df = pd.read_csv('RT_ratings.csv')
rt_df
```

```python
rotten_movies = rt_df['Title']
rt_to_imdb_score = []
for r in rotten_movies:
    imdb = f"imdb {r}"
    url = f'https://www.google.com/search?q={imdb}&rlz=1C1VDKB_enUS933US933&oq=imdb+batman+begins&aqs=chrome..69i57j0j0i22i30l2.11368j0j7&sourceid=chrome&ie=UTF-8'
    browser.visit(url)
    browser.click_link_by_partial_href('https://www.imdb.com/title/')
    html=browser.html
    soup=bs(html,'html.parser')
    sopa = soup.find('div',class_='TitleBlock__ButtonContainer-sc-1nlhx7j-3 fvllTh').get_text()
    rt_to_imdb_score.append(sopa)
    print(len(rt_to_imdb_score))
```

```python
rt_to_imdb_ratings = []
for rating in rt_to_imdb_score:
    rt_to_imdb_ratings.append(rating.split('/')[0])
rt_to_imdb_ratings
```

```python
true_score=[]
for score in rt_to_imdb_ratings:
    true_score.append(score.split('G')[1])
```

```python
score_df = pd.DataFrame(true_score)
score_df.columns=['IMDB score']
score_df
```

```python
#data = [movie[0],tomate_df[0],audience_df[0],imdb_score[1]]
#headers = ['Movie','Tomatometer','Audience Score','IMDB score']
#imdb_to_rt_final = pd.concat(data, axis=1, keys=headers)
#score_df
#rt_df
data = [rt_df,score_df]
final = pd.concat(data,axis=1)
```

```python
final = final[['Rank','RatingTomatometer','Title','No. of Reviews', 'IMDB score']]
final = final.rename(columns={'Rank':'IMDB Rank','Title':'Movie','RatingTomatometer':'Tomatometer'})
```

```python
final.to_csv('rt_to_imdb.csv',index=False)
```
