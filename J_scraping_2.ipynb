# Imports

```python
import pandas as pd 
from bs4 import BeautifulSoup as bs
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import nbconvert
import requests
import time
from splinter import Browser
```

# Create Browser

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

# Read in csv of RT 

```python
rt_df = pd.read_csv('RT_ratings.csv')
rt_df
```

# create for loop which uses google chrome to search for f"imdb {movie}
# click the first link which has www.imdb.com/title in it to find the exact movie 
# make browser a soup element and use find to grab the rating text
# append to list 

```python
rotten_movies = rt_df['Title']
rt_to_imdb_score = []
for r in rotten_movies:
    imdb = f"imdb {r}"
    url = f'https://www.google.com/search?q={imdb}&rlz=1C1VDKB_enUS933US933&oq=imdb+batman+begins&aqs=chrome..69i57j0j0i22i30l2.11368j0j7&sourceid=chrome&ie=UTF-8'
    browser.visit(url)
    browser.click_link_by_partial_href('https://www.imdb.com/title/')
    html=browser.html
    soup=bs(html,'html.parser')
    sopa = soup.find('div',class_='AggregateRatingButton__ContentWrap-sc-1il8omz-0 cMcGnJ').get_text()
    rt_to_imdb_score.append(sopa)
    print(len(rt_to_imdb_score))
```

```python
rt_to_imdb_score 
```

# use split to delete unnecessary text from 

```python
rt_to_imdb_ratings = []
for rating in rt_to_imdb_score:
    rt_to_imdb_ratings.append(rating.split('/')[0])
rt_to_imdb_ratings
```

# turn list into a dataframe for merge usage later and blank column 

```python
score_df = pd.DataFrame(rt_to_imdb_ratings)
score_df.columns=['IMDB score']
score_df
```

# create a dataframe from the original rotten tomato dataframe and add their IMDB score through concatenate 

```python
data = [rt_df,score_df]
final = pd.concat(data,axis=1)
final
```

# Remove unnamed, renmae columns and turn into final csv file

```python
final = final[['Rank','RatingTomatometer','Title','No. of Reviews', 'IMDB score']]
final = final.rename(columns={'Rank':'IMDB Rank','Title':'Movie','RatingTomatometer':'Tomatometer'})
```

```python
final.to_csv('rt_to_imdb.csv',index=False)
```
